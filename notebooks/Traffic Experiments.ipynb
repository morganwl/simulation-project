```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 2
B = 3
def time_gamma(t):
    return scipy.stats.gamma.pdf(t, A, B)
```

```python
x = np.linspace(1.5, 24, num=240)
plt.plot(x, time_gamma(x))
```

```python
def traffic_beta(p, d=1):
    p[p == 0] = np.finfo(np.float64).tiny
    alpha = p*d
    beta = (1 - p)*d
    return np.sqrt(np.random.beta(3,3, size=len(p)) * p)
    return np.random.beta(alpha, beta)
```

```python
plt.plot(x, traffic_beta(time_gamma(x), 1))
```
